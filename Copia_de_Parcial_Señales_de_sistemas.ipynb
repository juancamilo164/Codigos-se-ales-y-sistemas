La distancia media entre dos señales periódicas $x_1(t)\in\mathbb{R}$ y $x_2(t)\in\mathbb{R}$, se puede expresar a partir de la potencia media de la diferencia entre ellas:

$$
d(x_1, x_2) = \bar{P}_{x_1-x_2}=\lim_{T \to \infty} \frac{1}{T} \int_T |x_1(t)-x_2(t)|^{2}~dt.

```python
import numpy as np  #importar numpy para describir señales
import matplotlib.pyplot as plt   #se importa para generar graficas
from ipywidgets import interact, FloatSlider #se importa para generar el deslizante
from IPython.display import display #importa para reflejar cambion de amplitu

# Definir parámetros
omega = 1  # Velocidad angular
t = np.linspace(0, 10, 1000)  # Rango de tiempo

# Función para graficar x1(t) con el valor de A especificado
def plot_x1(A):
    # Calcular la parte real e imaginaria de la función x1(t)
    real_part = A * np.cos(omega * t)
    imaginary_part = A * np.sin(omega * t)

    # Graficar la función x1(t)
    plt.figure(figsize=(10, 5))
    plt.plot(t, real_part, label='Parte Real')
    plt.plot(t, imaginary_part, label='Parte Imaginaria')
    plt.xlabel('Tiempo t')
    plt.ylabel('x1(t)')
    plt.title('Gráfica de x1(t)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Crear un slider para el valor de A
slider = FloatSlider(min=0.1, max=5, step=0.1, value=1.0)
interact(plot_x1, A=slider)

# Mostrar el deslizante
display(slider)


```

## **Pregunta C**

Implemente una simulación para encontrar la salida del sistema lineal e invariante al tiempo $\mathcal{H}\{\cdot\}$, con respuesta al escalón $h_\epsilon[n]=\{2,4,1,5,0,10\}$, $n\in\mathbb{Z}$, ante la entrada análoga $x(t)=20(\cos(t/3)+\cos(t/4))$[A]. Amperios;Incluya los acondicionamientos necesarios de discretización y cuantización, asumiendo un microprocesador de 4 bits con entrada análoga de 4mA a 20 mA

Se aplica cero pendiente:

$\hat{x}(t)=m x(t)+ b$, con:

$m = \frac{\hat{x}_{max}(t)-\hat{x}_{min}(t)}{x_{max}(t)-x_{min}(t)}$

$b = \hat{x}_{min}(t)-m x_{min}(t)$

```python
#función cero y pendiente
def my_ceropend(x,xhmax=20,xhmin=4):
  #cálculo de pendiente y cero
  m = (xhmax-xhmin)/(x.max()-x.min())
  b = xhmin - m*x.min()
  xh = m*x+b
  return xh
```

## Explicacion del codigo
Calculo el cero pendiente con el minimo y maximo de entrada del sistema, con la funcion ceropen que recibe dos argumentos, despues de decribir, retonarmos el cero pendiente


Se discretiza la señal con numpy y se cuantiza con el numero de bits dado, con el maximo 20 y minimo 4, entrada analoga


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

T1 = 6*np.pi #perido señal 1
T2 = 8*np.pi #perido señal 2
F1 = 1/T1
F2 = 1/T2
T = 24*np.pi #perido cuasiperiódica
Fs = 10*max(F1,F2) #frecuencia muestreo
t = np.arange(0,3*T,1/Fs) #vector de tiempo, se muestran 3 periodos
x = 20*(np.cos(t/3)+np.cos(t/4))
xhmax = 20
xhmin = 4
xh = my_ceropend(x,xhmax, xhmin).reshape(-1,1)#aplicar cero y pendiente de 0 a 10
#asegurar vector columna

#cuantizar
nbits = 4 #número bits
Ne = 2**nbits #cantidad de estados
res = (xhmax-xhmin)/Ne #resolución cuantización
ve = np.arange(xhmin,xhmax,res).reshape(-1,1)#vector de estados
#asegurar vector columna

xq = np.zeros(len(xh)) #señal cuantizada
for i in range(len(xh)): #recorrer señal cero y pendiente
    xq[i] = ve[np.argmin(cdist(xh[i].reshape(-1,1),ve))] #se cálcula la distancia desde cada punto al vector de estados

#gráficas de señales
plt.plot(t,x,'r',label='$x(t)$')
plt.xlabel('t [s]')
plt.ylabel('Amplitud')
plt.show()
plt.plot(t,xh,'b',label='$\hat{x}(t)$')
plt.stem(t,xq,'y',label='$x_{q}[n]$')
plt.xlabel('n [muestra]')
plt.ylabel('Amplitud')
plt.legend()
plt.show()
```

##Explicacion Codigo
Se hace la definicion de parametros de la funcion dada, con nupy, ahi se establecen tambien los periodos y frecuencias correspondientes del modelo, se calcula el periodo de la funcion cuasiperiodica obtenida,establece la frecuencia de muestreo(Fs) como 10 veces la frecuencia maxima de F1,F2, se crea un vector de tiempo con el ts, tiempo de muestreo, con numpy arange, se genera la señal original con numpy y la funcion cos, Se aplica la funcion cero pendiente creada anteriormente, creando asi xh, se pasa a cuantizar en el numero de bits solicitado que en este caso es 4, se calcula cantidad de estados, se determina la cuantizacion y resolucion, se crea el vector con los estados discretos, se guarda la señal cuantizada y se pasa a graficar.


```python
he = np.array([2,4,1,5,0,10]) #respuesta al escalón del slit
h = np.diff(he)#respuesta al impulso del slit
print(h)#Arreglo con respuestas de impulso
```

La salida del sistema se obtiene mediante la operación de convolución:

$y[n]=x[n]*h[n]=\sum_{k=-\infty}^{\infty}x[k]h[n-k]$

## Explicacion del codigo
Se hace un arreglo numpy array para describir la entrada en respuesta escalon del sistema, h calcula entonces la salida en respuesta impulso

## Para cuantizar y descretizar la señal:

```python
y = np.convolve(xq,h,mode='same')#respuesta del slit -> se asegura tamaños iguales con mode=same
#graficar señal
plt.stem(t,y,'b',label='$y[n]$')
plt.legend()
plt.xlabel('n [muestra]')
plt.ylabel('Amplitud')
plt.show()
```
